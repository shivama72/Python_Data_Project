```python
#importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States']
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
US_data = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
US_data.reset_index(inplace=True)
US_data['month_no'] = pd.to_datetime(US_data['job_posted_month'], format='%B').dt.month
US_data.sort_values('month_no', inplace= True)
US_data.set_index('job_posted_month', inplace=True)
US_data.drop(columns=['month_no'], inplace=True)
```

```python
US_data
```

```python
US_other = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')
```

```python
US_merged = US_data.merge(US_other, on='job_posted_month')
```

```python
top_5 = (
    US_merged.sum()
    .sort_values(ascending=False)
    .head()
    .index
    .to_list()
    )

top_5
```

```python
US_merged[top_5].plot(kind='line')
plt.title('Monthly Job Postings for Top Tech Jobs in US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
