```python
#importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df.head()
```

```python
df['job_work_from_home'].value_counts().plot(kind = 'pie', startangle = 90, autopct = '%1.f%%')
plt.title('Work from home status')
plt.ylabel('')
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind = 'pie', autopct = '%1.f%%')
plt.title('Job Titles in my Dataset')
plt.ylabel('')
plt.show()
```

```python
df[['job_work_from_home', 'job_no_degree_mention', 'job_health_insurance']]
```

```python
fig, ax = plt.subplots(1,3)

dict_column = {
    'job_work_from_home': 'Work from Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'Health Insurance Offered'
}

for i,(column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle = 90, autopct = '%1.f%%', labels = ['False', 'True'])
    ax[i].set_title(title)

plt.show()
```
